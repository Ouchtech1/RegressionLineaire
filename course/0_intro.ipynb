# Introduction au Machine Learning 

## Définition du Machine Learning

Le **Machine Learning** ou **apprentissage automatique** est un sous-domaine de l'intelligence artificielle qui permet à des systèmes d'apprendre et de s'améliorer à partir de données sans être explicitement programmés pour chaque tâche. Contrairement aux méthodes déterministes et heuristiques, le Machine Learning repose sur des modèles qui ajustent leurs paramètres en fonction des données d'entrée pour faire des prédictions ou prendre des décisions.

### Exemple illustratif
Imaginez un programme qui doit prédire le prix d'une maison. Au lieu de coder des règles spécifiques (par exemple, "ajouter 10 000 € pour chaque chambre"), un modèle de Machine Learning peut apprendre ces relations à partir d'un ensemble de données sur les ventes passées.

En Machine Learning, la prédiction du prix d'une maison s'effectue à travers des étapes détaillées :

1. **Collecte des données**
   Collecter des données historiques incluant diverses caractéristiques des maisons vendues, comme le prix de vente, la superficie, le nombre de chambres, l'âge du bâtiment, la proximité des services, et la qualité des finitions.

2. **Entraînement du modèle**
   Appliquer des algorithmes pour détecter des corrélations entre les caractéristiques et le prix de vente. Le modèle pourrait apprendre, par exemple, que la proximité des transports en commun augmente le prix.

3. **Validation du modèle**
   Tester le modèle sur un ensemble de données séparé pour évaluer sa précision et ajuster les paramètres si nécessaire.

4. **Prédiction**
   Utiliser le modèle pour estimer le prix d'une maison en entrant ses caractéristiques, fournissant ainsi une évaluation basée sur les tendances actuelles du marché immobilier.

Le Machine Learning permet non seulement une estimation précise des prix mais aussi une adaptation aux évolutions du marché, offrant ainsi une supériorité en termes de flexibilité et de fiabilité par rapport aux méthodes déterministes et heuristiques.

### Distinction avec les heuristiques et méthodes déterministes

- **Méthodes déterministes** : solutions fixées et non adaptatives qui ne changent pas quelles que soient les données entrantes.
- **Heuristiques** : règles pratiques basées sur l'expérience, moins rigides que les méthodes déterministes mais ne s'adaptent pas automatiquement aux nouvelles données.

Le Machine Learning se distingue par sa capacité à apprendre les relations directement à partir des données, souvent avec une complexité que les approches déterministes et heuristiques ne peuvent pas facilement modéliser.










#### Méthodes déterministes
Un exemple de méthode déterministe pour estimer le prix d'une maison serait basé uniquement sur sa superficie :

$$Prix=2000€ \times \text{Superficie}_{m^2} $$

Dans cette formule, le prix est calculé exclusivement en fonction de la superficie, sans prendre en compte d'autres variables comme l'emplacement, l'âge du bâtiment, ou le nombre de chambres.

#### Heuristiques
Une approche heuristique pourrait incorporer plusieurs facteurs basés sur l'expérience :
- **Emplacement** : Ajouter 20 000 € si la maison est dans un quartier résidentiel.
- **Âge du bâtiment** : Soustraire 15 000 € si la maison a plus de 20 ans.
- **Nombre de chambres** : Ajouter 10 000 € pour chaque chambre au-delà de deux.

Cette méthode prend en compte plusieurs aspects de la propriété mais reste statique et ne s'adapte pas aux changements dynamiques du marché.


## La régression linéaire de Francis Galton

### Contexte
Francis Galton, un statisticien britannique du 19ème siècle, a observé que les traits physiques comme la taille des individus tendent à revenir vers une moyenne génération après génération. Cette observation a été fondamentale pour le développement initial des concepts de régression et de corrélation statistique.

### Le modèle de régression linéaire de Galton
Galton a utilisé la régression linéaire pour analyser la relation entre la taille des enfants et celle de leurs parents. Son modèle peut être formulé de la manière suivante :

$$y = a + bx$$

où :
- $y$ représente la taille de l'enfant,
- $x$ est la taille moyenne des parents,
- $a$ (l'intercept) représente la taille moyenne que l'on s'attend à ce que les enfants atteignent si la taille des parents est égale à la moyenne de la population,
- $b$ (la pente) mesure l'influence de la taille des parents sur celle de l'enfant.

### Hypothèses et conclusions
Galton supposait que si les parents sont plus grands ou plus petits que la moyenne, leurs enfants auront également tendance à être respectivement plus grands ou plus petits, mais moins extrêmes que leurs parents — un phénomène qu'il a appelé "régression vers la moyenne".

#### Hypothèses
1. **Linéarité** : La relation entre la taille des enfants et celle de leurs parents est linéaire.
2. **Indépendance** : Les tailles des différents enfants sont indépendantes les unes des autres.
3. **Homoscédasticité** : La variance des tailles des enfants est constante pour toutes les valeurs de taille des parents.

#### Conclusions
Galton a conclu que bien que les traits puissent être hérités, ils tendent à suivre une régression vers une moyenne populationnelle. Cette découverte a non seulement ouvert la voie à la génétique moderne mais a également posé les fondements de la régression linéaire comme outil statistique pour prédire une variable en fonction d'une autre.

### Impact en Machine Learning
Les travaux de Galton sur la régression linéaire ont montré comment on pouvait utiliser des données historiques pour prédire des issues futures, une idée qui est au cœur de nombreux algorithmes de Machine Learning. Ce principe est utilisé pour prédire des variables telles que les prix des maisons, les stocks en bourse, et bien d'autres phénomènes.

## Régression linéaire simple (approche statistique, économétrique et en tant que problème d'optimisation)
La régression linéaire est un exemple classique de modèle statistique utilisé pour prédire une variable quantitative à partir d'une autre. Le modèle est le suivant :

$$ y = \beta_0 + \beta_1 x + \epsilon $$

- $ \beta_0 $ et $\beta_1$ sont les paramètres du modèle,
- $\epsilon$ est le terme d'erreur (ou résidu)

```python
import numpy as np
import matplotlib.pyplot as plt

# Génération de données simulées
np.random.seed(0)
x = 2.5 * np.random.randn(100) + 1.5   # Array de 100 valeurs avec une moyenne de 1.5 et une variance de 2.5
res = 0.5 * np.random.randn(100)       # Générer 100 résidus 
y = 2 + 0.3 * x + res                  # Y dépend linéairement de x

# Tracé des données
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Données simulées')
plt.title('Régression linéaire simple')
plt.xlabel('Variable indépendante x')
plt.ylabel('Variable dépendante y')
plt.grid(True)
plt.legend()
plt.show()
```

### Approche statistique

L'approche statistique en Machine Learning consiste à utiliser des méthodes statistiques pour interpréter et modéliser les données. Cette approche est fondamentale pour comprendre les relations entre les variables et faire des prédictions.

**Nous allons utiliser numpy, une bibliothèque Python essentielle pour le calcul scientifique, qui offre des fonctionnalités robustes pour le traitement de tableaux et matrices, indispensables pour la manipulation des données et les calculs statistiques nécessaires à notre analyse.**

#### Équation de la Ligne

Une ligne dans un espace bidimensionnel peut être décrite par l'équation linéaire suivante :
$$ y = mx + c $$
Où :
- $y$ est la variable dépendante.
- $x$ est la variable indépendante.
- $m$ est la pente de la ligne.
- $c$ est l'ordonnée à l'origine.

#### Formulation Mathématique

##### Équations Normales

Les équations normales sont utilisées pour dériver les coefficients $m$ et $c$ directement à partir des données :
$$ \mathbf{A}^T\mathbf{A}\beta = \mathbf{A}^T\mathbf{y} $$
Où :
- $\mathbf{A}$ est une matrice des valeurs de la variable indépendante avec une colonne supplémentaire de uns (pour l'intercept),
- $\beta$ est le vecteur des paramètres $(m, c)$, ie $\beta=(m, c)=(\beta_0, \beta_1)$
- $\mathbf{y}$ est le vecteur des valeurs de la variable dépendante.

##### Solution

Résolution des équations normales :
$$ \beta = (\mathbf{A}^T\mathbf{A})^{-1}\mathbf{A}^T\mathbf{y} $$
Cette solution implique des opérations matricielles pour calculer les paramètres qui minimisent la somme des résidus au carré.

#### Calcul Pratique

Dans le code Python fourni, nous utilisons la fonction `np.linalg.lstsq`, qui calcule en interne :
$$ \beta = \text{np.linalg.lstsq}(\mathbf{A}, \mathbf{y})[0] $$
Cette fonction retourne la solution des moindres carrés à l'équation $\mathbf{A} \beta = \mathbf{y}$.

#### Représentation Visuelle

La ligne calculée $y = mx + c$ est ensuite tracée sur le graphique à points des données. La pente $m$ et l'interception $c$ de la ligne sont déterminées de manière à ce que la ligne s'ajuste au mieux à la distribution des points de données en termes de minimisation de la somme des distances verticales au carré entre les points de données et la ligne.

Ce cadre mathématique permet une modélisation efficace et efficace des relations linéaires au sein des données, fournissant à la fois une compréhension visuelle et quantitative claire de cette relation.

```python
# Calcul avec une vision statistique du problème
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]

# Graph des données et de la ligne de régression
plt.figure(figsize=(10, 5))
plt.plot(x, m*x + c, 'r', label='Ligne de régression')
plt.scatter(x, y, color='blue', label='Données simulées')
plt.title('Régression linéaire simple')
plt.xlabel('Variable indépendante x')
plt.ylabel('Variable dépendante y')
plt.grid(True)
plt.legend()
plt.show()
```

### Approche Économétrique avec Python

L'approche économétrique en régression analyse non seulement les relations entre variables mais aussi l'importance et la validité statistique de ces relations. Voici comment vous pouvez appliquer cette approche à notre ensemble de données simulé.

**Nous allons utiliser `statsmodels`, une bibliothèque Python qui fournit des classes et des fonctions pour l'estimation de nombreux modèles statistiques, ainsi que pour la conduite de tests statistiques et l'exploration de données.**


#### Interprétation du résumé

Le résumé fourni par `statsmodels` inclut plusieurs éléments clés :

- **Coef** : Les coefficients du modèle, où const est l'intercept et x1 la pente.
- **R-squared** : Le carré du coefficient de corrélation, qui mesure la qualité de l'ajustement du modèle.
- **P>|t|** : La valeur-p pour chaque coefficient testant l'hypothèse nulle que le coefficient est égal à zéro (pas d'effet).
- Visualisation de la régression
Pour visualiser les résultats de la régression, nous pouvons tracer la ligne de régression sur le graphique des données.


#### Visualisation de la régression
Pour visualiser les résultats de la régression, nous pouvons tracer la ligne de régression sur le graphique des données.

```python
import statsmodels.api as sm

# Ajouter une colonne de uns pour l'intercept
x_for_sm = sm.add_constant(x)

# Création du modèle
model = sm.OLS(y, x_for_sm)
results = model.fit()

# Afficher le résumé du modèle
print(results.summary())

# Prédiction à partir du modèle
y_pred = results.predict(x_for_sm)

# Tracé des données et de la ligne de régression
plt.figure(figsize=(10, 5))
plt.scatter(x_for_sm[:, 1], y, color='blue', label='Données simulées')
plt.plot(x_for_sm[:, 1], y_pred, 'r', label='Ligne de régression économétrique')
plt.title('Régression Linéaire Économétrique')
plt.xlabel('Variable indépendante x')
plt.ylabel('Variable dépendante y')
plt.legend()
plt.show()
```

### En tant que problème d'optimisation

L'approche d'optimisation vise à minimiser ou maximiser une fonction objectif, souvent à travers des méthodes numériques avancées. Pour notre problème de régression linéaire, nous utiliserons `scipy`, une bibliothèque qui propose des fonctions robustes pour la résolution de problèmes d'optimisation.


Nous allons utiliser `scipy.optimize` pour minimiser la somme des carrés des résidus, une méthode couramment utilisée dans la régression linéaire.


### Explications Mathématiques de l'Optimisation

La fonction que nous cherchons à minimiser dans le contexte de la régression linéaire est la **somme des carrés des résidus** (SSR). Elle est définie comme suit :
$$ \text{SSR} = \sum_{i=1}^n (y_i - (\beta_0 + \beta_1 x_i))^2 $$

où :
- $y_i$ sont les valeurs observées,
- $x_i$ sont les valeurs indépendantes,
- $\beta_0$ est l'intercept de la régression,
- $\beta_1$ est la pente de la régression.

Minimiser SSR revient à ajuster la ligne de régression de manière à ce que la somme des carrés des différences verticales entre les points de données observés et les points prédits par le modèle soit aussi petite que possible. Cela garantit que la ligne de régression représente au mieux la tendance des données.

```python
from scipy.optimize import minimize

# Fonction objectif : Somme des carrés des résidus
def objective_function(beta, x, y):
    return np.sum((y - (beta[0] + beta[1]*x))**2)

# Initialisation des paramètres (intercept et pente)
initial_guess = [0, 0]

# Utilisation de scipy pour minimiser la fonction objectif
result = minimize(objective_function, initial_guess, args=(x, y))

# Les paramètres optimisés
optimized_params = result.x

# Prédiction à partir des paramètres optimisés
y_pred = optimized_params[0] + optimized_params[1] * x

# Affichage des résultats
print("Optimized intercept:", optimized_params[0])
print("Optimized slope:", optimized_params[1])

# Tracé des données et de la ligne de régression optimisée
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Données simulées')
plt.plot(x, y_pred, 'r', label='Ligne de régression optimisée')
plt.title('Régression Linéaire par Optimisation')
plt.xlabel('Variable indépendante x')
plt.ylabel('Variable dépendante y')
plt.legend()
plt.show()
```
